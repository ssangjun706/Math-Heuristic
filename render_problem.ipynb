# 데이터 렌더링


```python
import os
import json
import pandas as pd
import random

from IPython.display import display, Markdown

DATA_DIR = os.path.join(os.getcwd(), "math-heuristics")
```

```python
model_name = "olmo-3-7b"
set_name = "hard"

result_file = os.path.join(
    DATA_DIR, set_name, f"{model_name}_math_perturb_{set_name}_result.json"
)

with open(result_file, "r", encoding="utf-8") as f:
    results = json.load(f)

results_df = pd.DataFrame(results)
```

```python
random_idx = random.randint(0, len(results_df) - 1)
item = results_df.iloc[random_idx]
display(Markdown(f"**랜덤 선택된 문제 ID:** {item['problem_id']}"))


if item is not None:
    display(Markdown(f"### 문제\n{item['problem']}"))
    display(Markdown(f"**정답:** {item['answer']}"))
    trials = item.get("trials", [])
    trial1 = next((t for t in trials if t.get("n", 0) == 1), None)
    if trial1:
        display(Markdown(f"**모델 생성 답변:** {trial1.get('generated_answer', '')}"))
        display(Markdown(f"**모델 생성 텍스트:**\n{trial1.get('generated_text', '')}"))
    else:
        display(Markdown("**trial=1 결과가 없습니다.**"))
```
